```python
import numpy as np

# how to generate a matrix (change entries and number of rows and columns for desired matrices)
M=np.array([
    [1,1,4],
    [0,2,0],
    [1,1,1]
])
print('Matrix M:', M, sep= '\n')

N=np.array([
    [4,0,1],
    [-2,3,5],
    [1,-1,0]
])
print('Matrix N:', N, sep= '\n')

# matrix addition and subtraction
print('Matrix M+N:', M+N, sep= '\n'     )

# matrix multiplication
print('Matrix MN:', M@N, sep= '\n'     )

# calculating determinant and trace
def matrix_properties(matrix):
    determinant = np.linalg.det(matrix)
    trace = np.trace(matrix)
    return determinant, trace

print('Determinant and Trace of M')
det, tr = matrix_properties(M)

print('Determinant:', det)
print('Trace:', tr)

# calcultating eigenvalues and eigenvectors
print(np.linalg.eig(M))
```

```python
# Question 1

import numpy as np

A=np.array([
    [1,1,4],
    [0,2,0],
    [1,1,1]
])
print('Matrix A:', A, sep= '\n')

print('Matrix A Eigs:', np.linalg.eig(A))

B=np.array([
    [-1,1,0],
    [1,2,1],
    [0,3,-1]
])
print('Matrix B:', B, sep= '\n')

print('Matrix B Eigs:', np.linalg.eig(B))

C=np.array([
    [-1,4,2],
    [4,-1,-2],
    [0,0,6]
])
print('Matrix C:', C, sep= '\n')

print('Matrix C Eigs:', np.linalg.eig(C))
```

```python
# Question 2 for Matrix A

import numpy as np

VA=np.array([
    [2,-2,5],
    [0,0,-3],
    [1,1,2]
])
print('Eig Vec Matrix V:', VA, sep= '\n')

DA=np.array([
    [3, 0, 0],
    [0, -1, 0],
    [0, 0, 2]
])

print('Eig Val Matrix D:', DA, sep= '\n')

print('V^-1:', np.linalg.inv(VA), sep= '\n')

print('VD(V^-1):', VA@DA@(np.linalg.inv(VA)), sep= '\n')

```

```python
# Question 2 for Matrix B

import numpy as np

VB=np.array([
    [1, -1, 1],
    [4, 0, -1],
    [3, 1,  3]
])
print('Eig Vec Matrix V:', VB, sep= '\n')

DB=np.array([
    [3, 0, 0],
    [0, -1, 0],
    [0, 0, -2]
])

print('Eig Val Matrix D:', DB, sep= '\n')

print('V^-1:', np.linalg.inv(VB), sep= '\n')

print('VD(V^-1):', np.round(VB@DB@(np.linalg.inv(VB))), sep= '\n')
```
