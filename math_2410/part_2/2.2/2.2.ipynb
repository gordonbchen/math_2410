```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")

# Circle.
def x(t):
    return np.cos(t)

def y(t):
    return 3 + np.sin(t)

# Ellipse.
def xe(t):
    return 2 * np.cos(t)

def ye(t):
    return 3 * np.sin(t)

ts = np.linspace(0, 2*np.pi, num=100)
xs = x(ts)
ys = y(ts)

xes = xe(ts)
yes = ye(ts)

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(xs, ys, label="circle")
ax.plot(xes, yes, label="ellipse")
ax.legend(loc=0)
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5);
```
