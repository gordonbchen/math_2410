# 2.1: Graphing 1D Functions

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8");


def f(x):
    return np.sin(x**2.0)

def g(x):
    return (x**3.0) + np.log(x)

def h(x):
    return np.sin(5.0*x) + np.log(x)*np.cos(x)

x_vals = np.linspace(1, 4, 20)

f_vals = f(x_vals)
g_vals = g(x_vals)
h_vals = h(x_vals)

plt.plot(x_vals, f_vals, label="f")
plt.plot(x_vals, g_vals, label="g")
plt.plot(x_vals, h_vals, label="h")

plt.legend(loc=0);
```

```python
xf = np.linspace(-5, 5, 200)
xg = np.linspace(0.1, 5, 200)
xh = np.linspace(0.1, 50, 300)

funcs = {
    "f": (xf, f(xf)),
    "g": (xg, g(xg)),
    "h": (xh, h(xh)),
}

fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(12, 14))
for ax, (name, (xs, ys)) in zip(axs, funcs.items()):
    ax.plot(xs, ys)
    ax.set_title(name);
```
