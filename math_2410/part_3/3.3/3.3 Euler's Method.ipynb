```python
# problem number 2
import numpy as np
import pandas as pd

# diff eq (change this!)
def f(t, y):
    if abs(t) < 1e-10:
        raise ValueError("Function is undefined near t = 0")
    return t * y**2 - y / t

# defines euler's method
def euler_method(x0, y0, h, x_end):
    t_values = [t0]
    y_values = [y0]
    t = t0
    y = y0

    while t < t_end - 1e-10:  # Adjusted condition to prevent overshooting
        y = y + h * f(t, y)
        t = round(t + h, 10)  # this is to avoid "floating-point" precision issues
        t_values.append(t)
        y_values.append(y)

    return t_values, y_values

# Initial conditions (change this!)
t0 = 1
y0 = 1
h = 0.2
t_end = 2

# tells it to compute values
t_vals, y_vals = euler_method(t0, y0, h, t_end)

# Create a table (data frame) for results
data = pd.DataFrame({'t': t_vals, 'y': y_vals})
print(data)

# Print the final value of y at t=(change this!)
print(f"y(2) ≈ {y_vals[-1]}")

# graph
import matplotlib.pyplot as plt

# data is from table
plt.plot(data['t'], data['y'])
plt.xlabel("t")
plt.ylabel("y")
plt.title("Euler's Method Approximation")
plt.show()
```
